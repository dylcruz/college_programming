```python
import math as m
import random as r

l = []

for i in range(0,1000):
    l.append(m.sqrt(r.randint(1,500)))
```

```python
l
```

```python
for i in range(0, len(l)):
    minIndex = i
    for j in range(i + 1, len(l)):
        if(l[minIndex] > l[j]):
            minIndex = j
    if(minIndex != i):
        temp = l[i]
        l[i] = l[minIndex]
        l[minIndex] = temp
```
